## Anuales

```python
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d})".format(pct, absolute)
# Grafico de torta
data_head, counts = np.unique(data['Período'], return_counts= True)
pl.title('Cantidad de transacciones por periodo')
_ = pl.pie(counts, labels=data_head, autopct=lambda pct: func(pct, counts))
```

```python
sns.histplot(data['Período'], kde=True)
_= pl.title('Cantidad de transacciones por periodo')
```

```python
# Grafico de torta
data_head, counts = np.unique(data['Denominación del objeto'], return_counts= True)
pl.title('Demanda porcentual de repuestos por sección anual')
_ = pl.pie(counts, labels=data_head, autopct='% .2f %%')
```
